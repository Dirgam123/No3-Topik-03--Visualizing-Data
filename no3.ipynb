```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('./Data.xlsx', sheet_name='Sheet1')
```

```python
negara = df['Negara']
penduduk = df['Jumlah Penduduk']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(negara, penduduk, marker='o', linestyle='-', color='r')
plt.title('Topik 03: Visualizing Data', fontsize=24)
plt.xlabel('Negara', fontsize=18)
plt.ylabel('Jumlah Penduduk', fontsize=18)
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
